```python



# Coletando informações do usuário
nome = input("Digite seu nome: ")  # Pede ao usuário para digitar seu nome
idade = input("Digite sua idade: ")  # Pede ao usuário para digitar sua idade
cidade = input("Digite a cidade onde você mora: ")  # Pede ao usuário para digitar a cidade onde mora

# Exibe uma mensagem de saudação ao usuário
print(f"\nOlá, {nome}! Vamos começar a análise de dados.")

# Importando bibliotecas necessárias
import pandas as pd  # Biblioteca para manipulação de dados
import matplotlib.pyplot as plt  # Biblioteca para criação de gráficos
import seaborn as sns  # Biblioteca para visualização de dados
from textblob import TextBlob  # Biblioteca para análise de sentimentos

# Gerando dados fictícios
data = {
    'produto': ['garrafa_ecologica', 'garrafa_plastica', 'bolsa_ecologica', 'bolsa_plastica', 'canudo_ecologico', 'canudo_plastico'],  # Nomes dos produtos
    'vendas': [1200, 1500, 800, 900, 700, 1100],  # Vendas de cada produto
    'menções_mídias_sociais': [
        'Amo minha garrafa ecológica! Ótima para o meio ambiente!',  # Menções nas mídias sociais sobre a garrafa ecológica
        'Garrafas de plástico são baratas, mas prejudiciais.',  # Menções sobre a garrafa de plástico
        'Bolsas ecológicas são as melhores!',  # Menções sobre a bolsa ecológica
        'Sacolas plásticas são convenientes, mas não são ecológicas.',  # Menções sobre a bolsa de plástico
        'Canudos ecológicos são fantásticos!',  # Menções sobre o canudo ecológico
        'Canudos de plástico estão em toda parte, muito ruim!'  # Menções sobre o canudo de plástico
    ]
}

# Criando um DataFrame a partir dos dados fictícios
df = pd.DataFrame(data)

# Exibindo o DataFrame para verificação
print("DataFrame inicial:")
print(df)

# Análise de Sentimentos
df['sentimento'] = df['menções_mídias_sociais'].apply(lambda x: TextBlob(x).sentiment.polarity)  # Calcula a polaridade do sentimento para cada menção

# Exibindo o DataFrame com os sentimentos calculados para verificação
print("\nDataFrame com sentimentos calculados:")
print(df)

# Visualização de dados de vendas
plt.figure(figsize=(10, 6))  # Define o tamanho da figura do gráfico
sns.barplot(x='produto', y='vendas', data=df)  # Cria um gráfico de barras para mostrar as vendas por produto
plt.title('Vendas de Produtos Sustentáveis vs Convencionais')  # Define o título do gráfico
plt.xlabel('Produto')  # Define o rótulo do eixo X
plt.ylabel('Vendas')  # Define o rótulo do eixo Y
plt.show()  # Exibe o gráfico

# Visualização de sentimentos
plt.figure(figsize=(10, 6))  # Define o tamanho da figura do gráfico
sns.barplot(x='produto', y='sentimento', data=df)  # Cria um gráfico de barras para mostrar a polaridade dos sentimentos por produto
plt.title('Análise de Sentimentos de Produtos Sustentáveis vs Convencionais')  # Define o título do gráfico
plt.xlabel('Produto')  # Define o rótulo do eixo X
plt.ylabel('Polaridade do Sentimento')  # Define o rótulo do eixo Y
plt.ylim(-1, 1)  # Define os limites do eixo Y para garantir que todos os sentimentos estejam visíveis
plt.show()  # Exibe o gráfico
```
